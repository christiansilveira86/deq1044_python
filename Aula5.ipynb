**Exemplo 1** <br>
Comecemos relembrando o exemplo introdutório que foi proposto no final da aula anterior, onde uma função $f(x) = x^2$ é calculada para um vetor $x$ variando de 0 a 5 e os resultados são gravados em diferentes variáveis: <br>

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,5.1,0.5,dtype=float)
f = np.zeros(x.size)
i = 0
while i < x.size:
    f[i] = x[i]**2
    i += 1

print(x.size)
print(x)
print(f)

plt.plot(x,f)
```

Nas linhas 1 e 2 importamos as bibliotecas _numpy_ e _matplotlib_, a primeira para realizar a criação dos vetores iniciais e a segunda para fazer a figura com os resultados obtidos. <br>
Na linha 3 criamos o vetor $x$ através da função _arange_ da biblioteca _numpy_. Assim, o vetor $x$ vai de 0 a 5 com incrementos de 0,5.  <br>
Na linha 4 iniciamos o vetor $f$ como um vetor de zeros do tamanho do vetor $x$. Esse vetor é criado para que possa receber, posteriormente, os valores reais calculados pela função. Essa criação prévia do vetor $f$ é o que se chama de pré-alocação de memória, o Python precisa que o vetor seja pré-alocado antes de realizarmos as operações na estrutura de repetição. <br>
Na linha 5 iniciamos o contador de iterações no 0. <br>
Na linha 6 iniciamos o laço de repetição _while_, que diz que enquanto $i$ for menor que o tamanho do vetor $x$, o bloco de código da estrutura deverá ser executado (linhas 7 e 8). $i$ não pode ser maior que _x.size_ porque assim estaríamos tentando acessar um valor inexistente do vetor $x$. Veja, por exemplo, este nosso caso: $x$ vai de 0 a 5 com incrementos de 0,5, assim o vetor $x$ possui 11 elementos (portanto _x.size_ é igual a 11) e não poderíamos acessar um décimo segundo elemento do vetor $x$. Assim, se $i$ fosse igual ou maior que 11 (_x.size_), tentaríamos calcular f(11) com x(11), mas como o décimo primeiro elemento de $x$ está na décima posição (lembre que o Python começa a contar do zero), este valor não irá existir. <br>
Na linha 7, calculamos a posição $i$ do vetor $f$ utilizando $x(i)$ (quando $x(0)=0$, $f(x)=0$; quando $x(1)=0.5$, $f(x)=0.25$; quando $x(2)=1.0$, $f(x)=1.0$, e assim or diante). <br>
Na linha 8, apenas incrementamos o valor de $i$ em uma unidade e repetimos o laço de iteração até que a condição da linha 6 não seja mais satisfeita. <br>
Nas linhas 10 à 11, usamos o comando para escrever na tela o tamanho do vetor $x$, os valores do vetor $x$ e os valores do vetor $f(x)$, respectivamente. <br>
Por fim, na linha 14, utilizamos o comando _plot(x,f)_ da biblioteca _matplotlib.pyplot_ para criar um gráfico com o vetor $x$ no eixo _x_ e o vetor $f(x)$ no eixo _y_. <br>

**Exemplo 2** <br>
Vejamos agora um outro caso. <br>
Por exemplo, podemos comparar a variação da capacidade calorífica de 3 gases diferentes com a temperatura, utilizando <br>
$$C_p/R = A + BT + CT^2$$ <br>
Vejamos uma comparação entre Propano, n-Butano e n-Pentano:

```python
# Importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
# Vetor de temperatura variando de 300 a 750 [K]
T = np.arange(300,751,5,dtype=float)
# Vetor do parâmetro A
A = np.array([1.213, 1.935, 2.464])
# Vetor do parâmetro B
B = np.array([28.785, 36.915, 45.351])*10**-3
# Vetor do parâmetro C
C = np.array([-8.824, -11.402, -14.111])*10**-6
Cp = np.zeros([3,T.size])
for i in [0, 1, 2]:
    for j in range(T.size):
        Cp[i,j] = A[i] + B[i]*T[j] + C[i]*T[j]**2;

plt.plot(T,Cp[0,:],label="Propano")
plt.plot(T,Cp[1,:],label="n-Butano")
plt.plot(T,Cp[2,:],label="n-Pentano")
plt.xlabel("Temperature [K]")
plt.ylabel("Cp/R")
plt.legend(loc="upper right")
plt.grid(color=[0.5, 0.5, 0.5], linestyle=':', linewidth=0.9)
```

No código acima calculamos o valor da capacidade calorífica $C_p/R$ de cada um dos 3 componentes para cada uma das temperaturas do vetor T (de 300 a 750 Kelvin) e depois plotamos um gráfico para visualizarmos os dados. <br>
Nas linhas 2 e 3 importamos as biblitoecas numérica e gráfica _numpy_ e _matplotlib.pyplot_, respectivamente. <br>
Na linha 5 criamos o vetor de Temperatura T, que vai de 300 a 750 K com incrementos de 5 K. E nas linhas 7, 9 e 11, criamos os vetores com os parâmetros de cada uma das espécies para o cálculo de seus $C_p$ (os dados foram retirados do Apêndice A.2 do livro Termodinâmica para Engenharia Química, Koretsky, 2007). <br>
Na linha 12, pré-alocamos a matriz que vai receber os valores das capacidades caloríficas de cada uma das espécies (por isso 3 linhas) para cada uma das temperaturas no vetor T (por isso o número de colunas é igual ao tamanho do vetor T (_T.size_)). <br>
Na linha 13, definimos com o _loop_ externo que a variável _i_ irá variar de 0 a 2. <br>
Na linha 14, definimos que o _loop_ interno irá de 0 até o número de elementos no vetor T. O uso da função _range()_ com apenas um argumento pressupõe que a contagem inicia do 0 e vai até o argumento (nesse caso, _T.size_) menos um - ou seja, se _T.size = 91_, _range(T.size)_ vai de 0 a 90. <br>
Assim, na linha 15, calculamos a capacidade calorífica para cada componente _i_ em cada temperatura _j_, utilizando o parâmetro do componente atual _i_ e a temperatura da iteração _j_. Por exemplo, a primeira vez que executamos a linha 15, _i=0_ e _j=0_, assim 
$$\frac {C_p(0,0)} {R} {=A(0)+B(0)*T(0)+C(0)*T(0)^2=1.213+28.785*300-8.824*300^2=-785523.287}$$ <br>
Da mesma maneira, a próxima execução da linha 15 será com _i=0_ e _j=1_, assim, a temperatura utilizada na equação acima não seria mais 300 K, mas 305 K, equivalente à posição [1] no vetor T. <br>
A linha 15 será executada até que _i_ e _j_ esgotem todos os valores (_i_ vai variar de 0 a 2 e _j_ de 300 a 750). <br>
Da linha 17 em diante, já fora dos dois _loops_, realizamos a _plotagem_ da figura. Dessa vez, utilizmos mais opções no comando _plot()_, como veremos. <br>
As linhas 17, 18 e 19 inserem as curvas dos gases Propano, n-Butano e n-Propano, respectivamente. O primeiro argumento é o próprio vetor de Temperatura, que será o eixo _x_ do nosso gráfico. O segundo argumento é a capacidade calorífica de cada um dos componentes (note que utilizamos as linhas identificadas, porque são elas que estão caracterizando cada espécie, e as colunas estão expressas como **_:_** porque utilizamos todos os valores calculados para cada temperatura). Incluímos, nestas 3 linhas, ainda um terceiro argumento "_label=" "_", que irá receber o nome da curva a ser plotado junto no gráfico. <br>
Os comandos _plt.xlabel_ e _plt.ylabel_ são utilizados para adicionar a legendas dos eixos _x_ e _y_, respectivamente. Já o comando _plt.legend_ é utilizados para definir algumas propriedades da legenda, neste caso utilizamos apenas para colocar a legenda no lado direito superior da figura, mais opções podem ser vistas em https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend. <br>
Por fim, na linha 23, adicionados as linhas de grade ao gráfico através do comando _plt.grid()_. A cor, neste caso, foi definida através da escala RGB, onde cada número representa uma graduação das cores primárias [R (vermelho, G (verde), B (azul)] (um exemplo diferente com a cor preta ('k') pode ser visto comentado na linha 24). O estilo da linha foi definido como :, o que resulta em vários pontos muito pequenos (experimente mudar para - ou --). E a espessura da linha foi definida como 0.9 (experimente modificar isto também). <br>

**Exemplo 3** <br>
    <div style="text-align: justify">
Suponha agora que você quer saber a que momento uma reação R atinge uma conversão X. <br>
Digamos que a reação seja de primeira ordem e conduzida em um reator em batelada. Assim o balanço molar de $A \rightarrow B$ é:
$$ \frac {dN_A}{dt} = -kC_AV = -kN_A $$
sendo $k = 0.0115 s^{-1}$.
Integrando a equação acima, obtemos: $\frac {N_A}{N_{A0}} = exp(-kt)$ <br>
A conversão de uma reação pode ser pensada como a quantidade de reagente que reagiu dividida pela quantidade de reagente alimentada. Assim:
    $$ X = \frac {N_{A0}-N_A}{N_{A0}} = 1 - \frac {N_A}{N_{A0}} = 1 - exp(-kt) $$
    Queremos saber em que momento a reação atinge 80%, 90%, 95% e 99% de conversão.
    
</div>

```python
# Importação das bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Declaração das variáveis
k = 0.0115; # Taxa de reação [s^-1]
t = np.arange(395) # Tempo [s]
X = np.zeros([t.size,1]) # Inicia o vetor que irá receber os valroes de conversão (uma linha e 101 colunas)
#X = 1 - np.exp(-k*t)
for i in t:
    X[i] = 1 - np.exp(-k*i)

X80 = np.nonzero(np.abs(0.80-X) < 1e-3)
print("Aos",X80[0],"segundos atingimos",X[X80[0]],"de conversão.")

# Repita o procedimento para as demais conversões

plt.plot(t,X)
plt.xlabel("Tempo [s]")
plt.ylabel("Conversão")
plt.grid(color='k',linestyle=":")
```

<div style="text-align: justify">
No caso de exemplo acima, existe uma maneira ainda mais simples de resolver o problema (linha 9 comentada), pois a operação se trata apenas de uma subtração de um esalar por um vetor. Mesmo assim, optamos por utilizar um laço de repetição para reforçarmos o funcionamento de _loops_.  <br>
É interessante notar o uso da função _nonzero_ da biblioteca _numpy_. A função _nonzero_ retorna os índices dos valores não-zeros de um vetor ou matriz: 
</div>

```python
a = np.array([0, 1, 2, 3, 0, 5, 6])
b = np.nonzero(a)
print(b)
```

Pode-se perceber que os elementos de índice 0 e 4 não foram impressos na resposta, pois estes são elementos iguais a 0. <br>
De outra maneira, podemos utilizar a função para encontrar valores que atinjam um certo critério. Por exemplo, selecionemos no vetor b apenas os valores menores que 3 e num vetor c os valores menores ou iguais a 3:

```python
a = np.array([0, 1, 2, 3, 0, 5, 6])
b = np.nonzero(a < 3)
c = np.nonzero(a <= 3)
print(b)
print(c)
```

O código retornou os índices 0, 1, 2 e 4 para o vetor b, que são todos elementos cujos valores são menores que 3. <br>
Por outro lado, o vetor c, de valores menores ou iguais a 3, retornou também, além dos elementos anteriores, o elemento 3, que é igual a 3. <br>
Assim, dessa maneira, na linha 13 do código que calcula a conversão ao longo do tempo, buscamos um valor que se distancie de 80% de convesão por um erro menor que 0,001.

**Exemplo 4** <br>
Um dos exemplos mais comuns que justifica o uso de estruturas de repetição é o cálculo de curvas de equilíbrio. Por exemplo, obtenhamos a curva de equilíbrio líquido-vapor (ELV) para a mistura binária de Acetona (1) + Água (2) utilizando as hipóteses de gas ideal e solução ideal a 393,15 K: <br>
$$y_iP=x_iP_{sat,i}$$
Assim, precisa-se calcular as pressões de saturação dos componentes 1 e 2 na temperatura de 393,15 K. Vejamos abaixo:


```python
# Importação de bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Número de componentes
NC = 2
# Temperatura
T = 393.15 # [K]
# Vetor das composições da fase líquida
xx = np.arange(0,1.01,0.01)
# Coeficientes de Antoine (cálculo da pressão de saturação)
A = np.array([16.6513, 18.3036])
B = np.array([2940.46, 3816.44])
C = np.array([-35.93, -46.13])
# Cria os vetores que irão receber valores
Psat = np.zeros(NC)
x = np.zeros(NC)
P = np.zeros(xx.size)
y = np.zeros((xx.size,NC))
# Cálculo das pressões de saturação
for i in range(NC):
    Psat[i] = np.exp(A[i] - B[i]/(T+C[i])) # mmHg

Psat = Psat*0.00133322 # Conversão de mmHg para bar
i=0 # Inicia o contador de iterações
for x1 in xx:
    x = np.array([x1, 1-x1])       # x1 é cada elemento do vetor xx, e x2 vem a ser 1-x1. Ambos são alocados no vetor temporário x
    P[i] = np.sum(np.dot(x,Psat))  # A pressão do sistema é dada pela média ponderada das pressões de saturação e frações molares (P = x1*P1sat + x2*P2sat)
    for j in range(NC):            # Para cada componente, calculamos um valor da fração molar na fase vapor
        y[i,j] = x[j]*Psat[j]/P[i] # y é uma matriz cujas linhas são correspondentes a cada valor de x e as colunas a cada espécie química
    i += 1                         # Aumenta-se em uma unidade o contador de iterações antes de repetir o laço

plt.plot(xx,P,label="Fração molar na fase líquida, $x_1$")
plt.plot(y[:,0],P,label="Fração nomar na fase vapor, $y_1$")
plt.legend(loc='best')
plt.xlabel("$x_1$,$y_1$")
plt.ylabel("Pressão [bar]")
plt.grid(color='k',linestyle=":")
```

O código acima inicia importando as bibliotecas necessárias. <br>
Depois, nas 6 e 8, definimos o número de componentes no sistema e a temperatura. <br>
Na linha 10 criamos o vetor $xx$, que irá de 0 a 1 com incrementos de 0,01. Os valores desse vetor serão utilizados como frações molares na fase líquida. <br>
Nas linhas 12, 13 e 14 definimos os vetores A, B e C, que contêm os valores dos parâmetros para a equação de Antoine (linha 22), que será utilizada para calcular a pressão de saturação de ada componente, dentro do _loop_ da linha 21. <br>
Nas linhas 16-19, pré-alocamos os espaços para as variáveis que serão calculadas. As pressões de saturação são apenas 2, tendo em vista que a temperatura é constante, sendo uma para cada espécie. $x$ receberá os valores de $x_1$ e $x_2$ em cada iteração, sendo assim um vetor de duas colunas, uma para cada componentes ($[x_1,x_2]$). $P$ é a pressão do sistema para cada par de valores de $x$, sendo assim, $P$ será um vetor do tamanho de $xx$. Por fim, $y$ receberá os valores das frações molares da fase vapor de cada componente (colunas, j) para cada pressão de equilíbrio do sistema (linhas, i). <br>
Assim, na linha 26 iniciamos as iterações para cada ponto da curva. Para cada iteração, pegamos um valor de $xx$ e alocamos em $x1$. Assim, o vetor $x$ é criado com o $x1$ atual e, portanto, com $x2$ atual ($1-x_1$). <br>
Na linha 28 calculamos a pressão para aquele par de frações molares como $x_1P_{1,sat}+x_2P_{2,sat}$, portanto, um somatório de produtos. Assim, _np.dot()_ realiza a multiplicação de $x_1P_{1,sat}$ e de $x_2P_{2,sat}$ e _np.sum()_ realiza a soma dos dois produtos. Assim, obtemos a pressão para aquele ponto. <br>
De posse da pressão, podemos calcula o par de valores das frações molares na fase vapor $y_1$ e $y_2$ para aquela pressão (variação em j, colunas, e i, linhas, respectivamente). <br>

Estamos sujeitos aos mais diferentes tipos de problemas e suas variações. Até mesmo os exemplos utilizados acima podem sofrer pequenas alterações que acarretariam uma lógica de programação diferente da utilizada. Cada caso é um caso diferente, cada programador também. Assim, cabe a você encontrar a maneira mais simples e eficiente de programar. <br>
Processos iterativos que utilizamos vetores e matrizes são complexos e muitas vezes de difícil compreensão. Espero que os exemplos utilizados tenham permitido a compreensão do uso dessas estruturas na lógica computacional. <br>
O exemplo acima utilizou a lei dos gases ideais e das soluções ideais para seus cálculos (por isso a curva da fase líquida é uma reta). Caso utilizássemos solução real para o problema, por exemplo van Laar, teríamos que contabilizar a não-idealidade através do coeficiente de atividade: <br>
$$y_iP=x_iP_{sat,i} \gamma_i$$ <br>
Lembrando que, para um sistema binário, van Laar fica na forma:<br>
 <br>
Assim, podemos reescrever o código:

```python
# Importação de bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Número de componentes
NC = 2
# Temperatura
T = 393.15 # [K]
# Vetor das composições da fase líquida
xx = np.arange(0.0,1.01,0.01)
# Coeficientes de Antoine (cálculo da pressão de saturação)
A = np.array([16.6513, 18.3036])
B = np.array([2940.46, 3816.44])
C = np.array([-35.93, -46.13])
# Parâmetros para van Laar (Acetona (1) + Água (2))
A12 = 2.1041
A21 = 1.5555
# Cria os vetores que irão receber valores
Psat = np.zeros(NC)
x = np.zeros(NC)
Pi = np.zeros(NC)
P = np.zeros(xx.size)
y = np.zeros((xx.size,NC))
# Cálculo das pressões de saturação
for i in range(NC):
    Psat[i] = np.exp(A[i] - B[i]/(T+C[i])) # mmHg

Psat = Psat*0.00133322 # Conversão de mmHg para bar
i = 0 # Inicia o contador de iterações
for x1 in xx:
    x = np.array([x1, 1-x1])       # x1 é cada elemento do vetor xx, e x2 vem a ser 1-x1. Ambos são alocados no vetor temporário x
    # Coeficientes de atividade calculados por van Laar
    gamma1 = np.exp(A12*((A21*x[1])/(A12*x[0]+A21*x[1]))**2)
    gamma2 = np.exp(A21*((A12*x[0])/(A12*x[0]+A21*x[1]))**2)
    gamma = [gamma1, gamma2]
    for j in range(NC):
        Pi[j] = x[j]*gamma[j]*Psat[j] # Pressão parcial de cada componente
    
    P[i] = np.sum(Pi) # Pressão do sistema é a soma das pressões parciais
    
    for j in range(NC):
        y[i,j] = Pi[j]/P[i]
    
    i += 1

# Figura
plt.plot(xx,P,label="Fração molar na fase líquida, $x_1$")
plt.plot(y[:,0],P,label="Fração nomar na fase vapor, $y_1$")
plt.legend(loc='best')
plt.xlabel("$x_1$,$y_1$")
plt.ylabel("Pressão [bar]")
plt.grid(color='k',linestyle=":")
```

Note que algumas pequenas mudanças na lógica do código foram feitar para que o código pudesse permanecer simples. <br>
O procedimento mais usual, no entanto, é criar uma função externa cuja única função seja calcular os coeficientes de atividade baseado nos dados de entrada. Veremos como criar e utilizar funções à partir da próxima aula. <br>
